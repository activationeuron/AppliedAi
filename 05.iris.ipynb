## PDF and CDF for pettal length 

```python
counts,bin_edges=np.histogram(iris["petal_length"],bins=10,density=True)
```

```python
pdf=counts/(sum(counts))
```

```python
pdf
```

```python
bin_edges
```

```python
cdf=np.cumsum(pdf)
```

```python
plt.plot(bin_edges[1:],pdf) #wrong 9,7 redo
plt.plot(bin_edges[1:],cdf)
plt.grid()
```
