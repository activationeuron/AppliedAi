```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
iris=pd.read_csv("./05.iris.csv")
```

```python

sns.boxplot(x="species",y="petal_length",data=iris,)
plt.show()
```

```python
sns.violinplot(x="species",y="petal_length",data=iris)
```

```python

sns.jointplot(x="petal_width",y="petal_length",data=iris,kind="kde")
```
