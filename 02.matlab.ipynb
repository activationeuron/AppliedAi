```python
import numpy as np
```

```python
t=np.arange(0,5,0.2)
```

```python
plt.plot(t,t**2,"b--",label="^2")
plt.plot(t,t**2.3,"g^",label="^2.3")
plt.plot(t,t**2.5,"ks",label="^2.5")
plt.legend()
plt.grid()
plt.show()
```

```python
t1=np.arange(0.2,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)
def f(t):
    return np.exp(-2)*np.cos(2*np.pi*t)
```

```python
plt.figure(1)
plt.subplot(211) #two rows, ccolums one, figure one 
plt.grid()
plt.plot(t1,f(t1),"b-")
plt.subplot(212) #two rows, columns one, figure two
plt.plot(t1,f(t1),"r--")
plt.show()
```
